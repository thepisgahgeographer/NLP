```python
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor

#connect to the tabular data
fire_path = r"C:\Users\yuri7100\Desktop\Fires.csv"
data = pd.read_csv(fire_path)
data.describe()
data.columns

#Create a y variable or the to-be predicted target
y = data.STAT_CAUSE_CODE

#Create features 
features = ["DISCOVERY_DOY","FIRE_YEAR"]
X = data[features]
X.describe()

#Build Regressor
model = DecisionTreeRegressor()
model.fit(X,y)
prediction = model.predict(X.head())

predicted_fires = model.predict(X)
predicted_fires_final = mean_absolute_error(y, predicted_fires)
predicted_fires_final
```
